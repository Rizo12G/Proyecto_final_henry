```python
import pandas as pd 
import pyarrow as pa
import pyarrow.parquet as pq
import os
```

```python
df_taxis = pd.read_parquet('../Proyecto_final_henry/raw_data/taxis/taxis.parquet')
```

```python
# Define una lista para almacenar los trimestres y el rango de meses correspondientes
trimestres = [(1, 'T1'), (4, 'T2'), (7, 'T3'), (10, 'T4')]

# Recorre cada trimestre y filtra los datos correspondientes
for mes_inicio, trimestre in trimestres:
    # Filtra el DataFrame por trimestre usando la funci√≥n `dt.quarter`
    df_trimestre = df_taxis[df_taxis['tpep_pickup_datetime'].dt.month.isin(range(mes_inicio, mes_inicio + 3))]
    
    # Guarda cada segmento trimestral en un archivo `.parquet`
    nombre_archivo = f'../Proyecto_final_henry/raw_data/taxis/taxis_{trimestre}.parquet'
    df_trimestre.to_parquet(nombre_archivo, index=False)
    print(f"Archivo guardado: {nombre_archivo}")
```

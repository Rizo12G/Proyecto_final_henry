```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
taxis = pd.read_csv('../clean_data/taxis_clean.csv')
```

```python
#Convertimos las columnas de fechas a datetime y verificamos la duración del viaje 
taxis['pickup_datetime'] = pd.to_datetime(taxis['pickup_datetime'])
taxis['dropoff_datetime'] = pd.to_datetime(taxis['dropoff_datetime'])
```

```python
# Duración del viaje (en minutos)
taxis['calculated_duration'] = (taxis['dropoff_datetime'] - taxis['pickup_datetime']).dt.total_seconds() / 60

#Comparación duracion calculada con columna 'duration'
taxis['duration_diff'] = taxis['calculated_duration'] - taxis['duration']
```

Análisis Univariable 

```python
# Distribución de trip_distance
sns.histplot(taxis['trip_distance'], bins=50, kde=True)
plt.title('Distribución de distancia de viajes')
plt.xlabel('Distancia (millas)')
plt.ylabel('Frecuencia')
plt.show()

# Distribución de fare_amount
sns.boxplot(x=taxis['fare_amount'])
plt.title('Distribución de tarifas')
plt.show()

# Análisis de tipos de viaje
sns.countplot(x='type', data=taxis)
plt.title('Conteo de tipos de viaje')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(taxis['trip_distance'], bins=30, kde=False, color='blue')
plt.title('Distribución de la Distancia de los Viajes')
plt.xlabel('Distancia (millas)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(taxis['fare_amount'], bins=50, kde=True, color='green')
plt.title('Distribución de Tarifas')
plt.xlabel('Tarifa ($)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='payment_type', y='tip_amount', data=taxis, ci=None, palette='coolwarm')
plt.title('Promedio de Propinas por Tipo de Pago')
plt.xlabel('Tipo de Pago')
plt.ylabel('Propina Promedio ($)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x=taxis['type'], y=taxis['duration'], jitter=True, alpha=0.5, palette='Set1')
plt.title('Duración de los Viajes por Tipo')
plt.xlabel('Tipo de Viaje')
plt.ylabel('Duración (minutos)')
plt.show()
```

Análisis Bivariado 

```python
#Relación entre distancia y tarifa 
plt.figure(figsize=(8, 6))
plt.hexbin(taxis['trip_distance'], taxis['fare_amount'], gridsize=50, cmap='Blues', mincnt=1)
plt.colorbar(label='Frecuencia')
plt.title('Distancia vs Tarifa (Hexbin)')
plt.xlabel('Distancia (millas)')
plt.ylabel('Tarifa ($)')
plt.show()
```

```python
#Relacion entre tarifa y propina 
plt.figure(figsize=(8, 6))
plt.hexbin(taxis['fare_amount'], taxis['tip_amount'], gridsize=50, cmap='Purples', mincnt=1)
plt.colorbar(label='Frecuencia')
plt.title('Tarifa vs Propina (Hexbin)')
plt.xlabel('Tarifa ($)')
plt.ylabel('Propina ($)')
plt.show()

taxis['fare_range'] = pd.cut(taxis['fare_amount'], bins=[0, 10, 20, 30, 50, 100, 200], labels=['0-10', '10-20', '20-30', '30-50', '50-100', '100+'])

plt.figure(figsize=(10, 6))
sns.boxplot(x='fare_range', y='tip_amount', data=taxis)
plt.title('Distribución de propinas por rango de tarifa')
plt.xlabel('Rango de tarifa ($)')
plt.ylabel('Propina ($)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='payment_type', y='tip_amount', data=taxis, palette='pastel')
plt.title('Distribución de Propinas por Tipo de Pago')
plt.xlabel('Tipo de Pago')
plt.ylabel('Propina ($)')
plt.show()
```

Análisis Multivariable 

```python
# Crear rango de tarifas para agrupar
taxis['fare_range'] = pd.cut(taxis['fare_amount'], bins=[0, 10, 20, 30, 50, 100, 200], labels=['0-10', '10-20', '20-30', '30-50', '50-100', '100+'])

# Boxplot: Propina por rango de tarifa y tipo de viaje
plt.figure(figsize=(10, 6))
sns.boxplot(x='fare_range', y='tip_amount', hue='type', data=taxis)
plt.title('Propina por Rango de Tarifa y Tipo de Viaje')
plt.xlabel('Rango de Tarifa ($)')
plt.ylabel('Propina ($)')
plt.legend(title='Tipo de Viaje')
plt.show()
```

```python
# Crear rangos de tarifas
taxis['fare_range'] = pd.cut(taxis['fare_amount'], bins=[0, 10, 20, 30, 50, 100], labels=['0-10', '10-20', '20-30', '30-50', '50-100'])

# Gráfico Violinplot
plt.figure(figsize=(12, 6))
sns.violinplot(x='fare_range', y='tip_amount', hue='type', data=taxis, split=True, palette='muted')
plt.title('Distribución de Propinas por Rango de Tarifa y Tipo de Viaje')
plt.xlabel('Rango de Tarifa ($)')
plt.ylabel('Propina ($)')
plt.legend(title='Tipo de Viaje')
plt.show()
```

```python
# Calcular correlación y mostrar heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = taxis[['trip_distance', 'fare_amount', 'tip_amount', 'duration', 'total_amount']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Correlación entre Variables Numéricas')
plt.show()
```

```python
# Agregar columna del día de la semana
taxis['pickup_day'] = taxis['pickup_datetime'].dt.day_name()

# Gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='pickup_day', y='duration', data=taxis, ci=None, color='skyblue', estimator=np.mean)
plt.title('Duración Promedio del Viaje por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Duración Promedio (minutos)')
plt.show()

# Tarifa por día de la semana
plt.figure(figsize=(12, 6))
sns.barplot(x='pickup_day', y='fare_amount', data=taxis, ci=None, color='orange', estimator=np.mean)
plt.title('Tarifa Promedio por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Tarifa Promedio ($)')
plt.show()
```

Análisis Temporal 

```python
# Crear columna para el día de la semana
taxis['pickup_day'] = taxis['pickup_datetime'].dt.day_name()

# Análisis de tarifas por día
sns.boxplot(x='pickup_day', y='fare_amount', data=taxis, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Distribución de tarifas por día de la semana')
plt.xticks(rotation=45)
plt.show()

# Análisis de viajes por hora
taxis['pickup_hour'] = taxis['pickup_datetime'].dt.hour
sns.countplot(x='pickup_hour', data=taxis)
plt.title('Número de viajes por hora del día')
plt.show()
```

Segmentación por Ubicación 

```python
# Conteo de ubicaciones de origen y destino
top_pickup = taxis['PULocationID'].value_counts().head(10)
print("Top 10 ubicaciones de recogida:\n", top_pickup)

# Visualización
sns.barplot(x=top_pickup.index, y=top_pickup.values)
plt.title("Top 10 ubicaciones de recogida")
plt.xlabel("ID de ubicación")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
heatmap_data = taxis.pivot_table(index='PULocationID', columns='DOLocationID', aggfunc='size', fill_value=0)
sns.heatmap(heatmap_data, cmap='Blues', cbar_kws={'label': 'Frecuencia'})
plt.title('Frecuencia de Viajes por Ubicaciones (Origen-Destino)')
plt.xlabel('ID de Ubicación de Destino')
plt.ylabel('ID de Ubicación de Recogida')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
avg_duration = taxis.groupby('PULocationID')['duration'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=avg_duration.index, y=avg_duration.values, palette='viridis')
plt.title('Top 10 Ubicaciones de Recogida con Mayor Duración Promedio')
plt.xlabel('ID de Ubicación de Recogida')
plt.ylabel('Duración Promedio (minutos)')
plt.show()
```
